```python
x = [({i}, [[i]] * i, {i: (*range(i * 3000),)}) for i in range(99)]
```

```python
from copy import deepcopy
```

```python
import copium.patch; copium.patch.disable()  # in case copium-autopatch is installed
```

```python
stdlib = %timeit -o deepcopy(x)
```

```python
import copium.patch; copium.patch.enable()
```

```python
copium = %timeit -o deepcopy(x)
```

```python
import json
from pathlib import Path

def timeit_to_dict(result):
    return {
        "average": result.average,
        "stdev": result.stdev,
        "best": result.best,
        "worst": result.worst,
        "loops": result.loops,
        "repeat": result.repeat,
        "timings": result.timings,
    }

results = {"stdlib": timeit_to_dict(stdlib), "copium": timeit_to_dict(copium)}

chart_json_file = Path(".bench/chart_results.json")
chart_json_file.parent.mkdir(exist_ok=True)
chart_json_file.write_text(json.dumps(results, indent=2))
print(f"Saved: stdlib={stdlib.average:.4f}s, copium={copium.average:.4f}s, speedup={stdlib.average / copium.average:.1f}x")
```
